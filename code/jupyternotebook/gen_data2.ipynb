```python
import pandas as pd
import numpy as np
from sklearn.utils import shuffle
```

```python
benign = pd.read_csv("../../data/benign.csv")
dga = pd.read_csv("../../data/dga-feed.csv")
```

```python
benign["domain_type"] = "benign"
benign = benign.sample(30000)
```

```python
dga["domain_type"] = "dga"
dga = dga.sample(30000)
```

```python
data = [benign, dga]
data = pd.concat(data)
data = shuffle(data)
data = data.reset_index(drop=True)
data
```

```python
data.to_csv('expanded_data.csv',index=False)
```
